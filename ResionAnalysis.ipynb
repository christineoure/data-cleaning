```python
# Separate numerical columns
numerical_cols = data_df.select_dtypes(include=['number']).columns.tolist()

# Separate categorical columns
categorical_cols = data_df.select_dtypes(exclude=['number']).columns.tolist()

# Display the results
print("Numerical Columns:")
print(numerical_cols)

print("\nCategorical Columns:")
print(categorical_cols)
```

```python
# Check for missing values
print("\nMissing values in each column:")
data_df.isnull().sum()
```

```python
# Fill missing values in numerical columns with the mean
data_df[numerical_cols] = data_df[numerical_cols].fillna(data_df[numerical_cols].mean())

# Fill missing values in categorical columns with the mode
for col in categorical_cols:
    data_df[col] = data_df[col].fillna(data_df[col].mode()[0])

# Verify that there are no missing values
print("\nMissing values after filling:")
data_df.isnull().sum()
```

```python
# Check for missing values
print("\nMissing values in each column:")
data_df.isnull().sum()
```

```python
# Check for duplicate rows
duplicate_rows = data_df.duplicated().sum()
print(f"Number of duplicate rows: {duplicate_rows}")
```

```python
data_df.dtypes
```

```python
# Convert numerical columns that may be incorrectly recognized as object types
for col in data_df.columns:
    # If the column should be numeric but is an object, convert it
    if data_df[col].dtype == 'object':
        try:
            data_df[col] = pd.to_numeric(data_df[col].str.replace(',', ''), errors='raise')
        except ValueError:
            pass  # If conversion fails, keep it as an object
```

```python
# Convert categorical columns to 'category' data type
categorical_cols = data_df.select_dtypes(exclude=['number']).columns
for col in categorical_cols:
    data_df[col] = data_df[col].astype('category')
```

## **3. Feature Engineering**

### **3.1 Creating New Features**
- Describe the new features you created and the rationale behind them. For example:
  - "New features such as 'CO2 per GDP' and 'Cumulative CO2 including Land Use Change' were created to capture the relationship between economic output and emissions more effectively."

### **3.2 Feature Selection**
- Explain any feature selection methods used. For example:
  - "Features with low variance or high correlation with other features were dropped to improve model performance."




```python
#CO2 Emissions Estimate
data_df['CO2_Emissions_Estimate'] = data_df['primary_energy_consumption'] * 0.3
```

```python
#Energy Consumption per Capita
data_df['Energy_per_Capita'] = data_df['primary_energy_consumption'] / data_df['population']
```

```python
#CO2 Intensity (CO2 per unit of GDP)
data_df['CO2_Intensity'] = data_df['co2'] / data_df['gdp']
```

```python
# Display the first few rows to check the new features
data_df[['CO2_Emissions_Estimate', 'Energy_per_Capita', 'CO2_Intensity']].head()
```

```python
#Polynomial Features
from sklearn.preprocessing import PolynomialFeatures

# Select relevant numerical features for interaction
numerical_features = data_df[['primary_energy_consumption', 'gdp', 'population']]

# Generate polynomial features (e.g., degree 2 for quadratic terms)
poly = PolynomialFeatures(degree=2, include_bias=False)
poly_features = poly.fit_transform(numerical_features)

# Convert back to DataFrame with appropriate feature names
poly_feature_names = poly.get_feature_names_out(numerical_features.columns)
poly_df = pd.DataFrame(poly_features, columns=poly_feature_names)

# Add polynomial features to the original DataFrame
data_df = pd.concat([data_df, poly_df], axis=1)
```

```python
#Logarithmic Transformations
import numpy as np

# Apply logarithmic transformation to relevant features
data_df['log_gdp'] = np.log1p(data_df['gdp'])  # log1p handles log(0) by computing log(1 + x)
data_df['log_population'] = np.log1p(data_df['population'])
data_df['log_energy_consumption'] = np.log1p(data_df['primary_energy_consumption'])
```

```python
#Categorical Feature Encoding (One-Hot Encoding)
# Select categorical columns for encoding
categorical_columns = ['iso_code']  # Example categorical column

# Apply one-hot encoding
data_df = pd.get_dummies(data_df, columns=categorical_columns, drop_first=True)
```

```python
#Interaction Terms
# Create interaction terms manually (e.g., between energy consumption and GDP)
data_df['energy_gdp_interaction'] = data_df['primary_energy_consumption'] * data_df['gdp']
data_df['gdp_population_interaction'] = data_df['gdp'] * data_df['population']
```

```python
# Combine polynomial features with the original DataFrame
data_df = pd.concat([data_df, poly_df], axis=1)
```

```python
# Display the first few rows to check the new features
data_df.head()
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, r2_score

# Assuming 'data_df' is your DataFrame and 'co2' is your target variable

# Define the feature matrix (X) and the target vector (y)
X = data_df.drop(columns=['co2'])  # Features: all columns except the target
y = data_df['co2']  # Target: CO2 emissions

# Split the dataset into training and testing sets (70/30 split)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# Initialize and train the linear regression model
linear_regressor = LinearRegression()
linear_regressor.fit(X_train, y_train)

# Predict and evaluate
y_pred_lr = linear_regressor.predict(X_test)
print("Linear Regression - RMSE:", mean_squared_error(y_test, y_pred_lr, squared=False))
print("Linear Regression - R^2 Score:", r2_score(y_test, y_pred_lr))
```

```python
# Initialize and train the decision tree regression model
tree_regressor = DecisionTreeRegressor(random_state=42)
tree_regressor.fit(X_train, y_train)

# Predict and evaluate
y_pred_tree = tree_regressor.predict(X_test)
print("Decision Tree Regression - RMSE:", mean_squared_error(y_test, y_pred_tree, squared=False))
print("Decision Tree Regression - R^2 Score:", r2_score(y_test, y_pred_tree))
```

```python
# Initialize and train the random forest regression model
forest_regressor = RandomForestRegressor(random_state=42, n_estimators=100)
forest_regressor.fit(X_train, y_train)

# Predict and evaluate
y_pred_forest = forest_regressor.predict(X_test)
print("Random Forest Regression - RMSE:", mean_squared_error(y_test, y_pred_forest, squared=False))
print("Random Forest Regression - R^2 Score:", r2_score(y_test, y_pred_forest))
```

```python
# Initialize and train the support vector regression model
svr_regressor = SVR(kernel='rbf')
svr_regressor.fit(X_train, y_train)

# Predict and evaluate
y_pred_svr = svr_regressor.predict(X_test)
print("Support Vector Regression - RMSE:", mean_squared_error(y_test, y_pred_svr, squared=False))
print("Support Vector Regression - R^2 Score:", r2_score(y_test, y_pred_svr))
```

```python
# Initialize and train the gradient boosting regression model
gboost_regressor = GradientBoostingRegressor(random_state=42)
gboost_regressor.fit(X_train, y_train)

# Predict and evaluate
y_pred_gboost = gboost_regressor.predict(X_test)
print("Gradient Boosting Regression - RMSE:", mean_squared_error(y_test, y_pred_gboost, squared=False))
print("Gradient Boosting Regression - R^2 Score:", r2_score(y_test, y_pred_gboost))
```
